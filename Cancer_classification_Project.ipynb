First we start by importing the libraries we need.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

next we download the dataset we would be using. we can get if from kaggle or from the sklearn library.

```python
import sklearn.datasets
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and processing


```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_dataset
```

Next we'll have to load this data to a dataframe. The data exist in as an array which does not look nice to me.

```python
# load the data to a dataframe.
df = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows to have a sneak peek at it.
df.head()
```

```python
# now we want to add a target column to our dataframe.
df['label'] =  breast_cancer_dataset.target
```

```python
df.tail()
```

```python
# let's get the dimension of the data
df.shape
```

```python
# get some information about the data.
df.info()
```

```python
# let's check for missing  values
df.isnull().sum()
```

```python
df.describe()
```

```python
# now i want to find out how many malignant and benign causes we have in our data
df.label.value_counts()
```

>> 0 - Malignant that is harmful
>> 1 - Benign that is not harmful

```python
# let's do a  groupby to get some specific summary about our data
df.groupby('label').mean() 
```

One thing we can see from the table above is that the values for most of the malignant cases are higher than the ones for the benign cases.

```python
plt.figure(figsize=(26,20))
sns.heatmap(df.corr(), annot=True, cmap="winter")
```

```python
# we need to seperate our target from the feature values.
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
# let's see what our features look like
X
```

```python
# let's see what our labels look like.
Y
```

Next split the data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
# now i'm interested in seeing the dimension of the data.
print(X.shape, X_train.shape, X_test.shape)
```

Now can proceed to training our model. 

```python
# first load an instance of the model.
model = LogisticRegression()
```

```python
# train the logistic model using the training data we had initially splitted.
model.fit(X_train, Y_train)
```

Model Evaluation. Right now we have successfully built the model, the next thing we have to be conscious of is the evaluation of our model. This will tell us how good or bad our model is.

First we will evaluate the model based on the X_train data. i will then us it to make predictions as to what would be the Y_train values. Then i would use this new value to compare with the actual Y_train values. This will help me measure the accuracy on the trained data. Then i will proceed to the test data and do  the same thing.

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
# now, let's print and see how well our model did with prediction
print( 'the accuracy of the training data is', f"{training_data_accuracy:.2f}")
```

Now i want to do the same thing with the test data. We want to see how well our model will do with unseen data.

```python
# accuracy on the test data.
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
# now, let's see how well our model does with prediction on the test_data
print( 'the accuracy of the test data is', f"{test_data_accuracy:.2f}")
```

Next we will build a predictive system

```python
input_data = (12.45,15.7,82.57,477.1,0.1278,0.17,0.1578,0.08089,0.2087,0.07613,0.3345,0.8902,2.217,27.19,0.00751,0.03345,0.03672,0.01137,0.02165,0.005082,15.47,23.75,103.4,741.6,0.1791,0.5249,0.5355,0.1741,0.3985,0.1244)

# change the input data to a numpy array
input_data_as_numpy_array = np.array(input_data)

# reshape this data.
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

# make prediction
prediction = model.predict(input_data_reshape)

print(prediction)

if prediction[0] == 0:
  print('The Breast Cancer is Malignant')

else:
  print('The Breast Cancer is Benign')  
```
